```python
for k in url:      #fetching data of various flights from each page
    driver.get(k)   
    time.sleep(2)
    
    # Extracting Airline name from xpath
    airline = driver.find_elements_by_xpath("//div[@class='fs-13 airline-name no-pad col-8']/span")
    for j in airline:
        airline_name.append(j.text)   

    time.sleep(1)
    
    # Extracting source from xpath
    sources = driver.find_elements_by_xpath("//div[@class='i-b col-4 no-wrap text-right dtime col-3']/p")      
    for j in sources:
        source.append(j.text)   

    time.sleep(1)
    
    # Extracting destination from xpath
    destinations = driver.find_elements_by_xpath("//div[@class='i-b pdd-0 text-left atime col-5']/p[2]")      
    for j in destinations:
        destination.append(j.text)
      
    time.sleep(1)
    
    # Extracting departure time from xpath
    deptime = driver.find_elements_by_xpath("//div[@class='i-b pr']")      
    for j in deptime:
        departure_time.append(j.text)   

    time.sleep(1)
    
    # Extracting arrival time from xpath
    arrtime = driver.find_elements_by_xpath("//div[@class='i-b pdd-0 text-left atime col-5']/p[1]")      
    for j in arrtime:
        arrival_time.append(j.text)
    
    time.sleep(1)
    
    # Extracting duration from xpath
    durations = driver.find_elements_by_xpath("//p[@class='fs-12 bold du mb-2']")      
    for j in durations:
        duration.append(j.text)
    
    time.sleep(1)
    
    # Extracting total stops from xpath
    stops = driver.find_elements_by_xpath("//div[@class=' font-lightgrey fs-10 tipsy i-b fs-10']/span")      
    for j in stops:
        total_stops.append(j.text)

    time.sleep(1)
    
    # Extracting price of flights from xpath
    pri = driver.find_elements_by_xpath("//div[@class='i-b tipsy fare-summary-tooltip fs-18']")      
    for j in pri:
        price.append(j.text)
        
    time.sleep(1)
    
    # Extracting Date of Journey from xpath
    date = driver.find_elements_by_xpath("//div[@class='bold fs-12 mb-5']")      
    for j in date:
        date_journey.append(j.text)
        
    time.sleep(1)
    
    # Extracting meal info from xpath
    info = driver.find_elements_by_xpath("//div[@class='fs-12 no-wrap']/span[2]/text()")      
    for j in info:
        meal.append(j.text)

    time.sleep(1)
```
